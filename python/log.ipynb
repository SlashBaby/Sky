```python
import json
```

```python
def read_json(filename):
    with open(filename) as f:
        data = json.load(f)
        return data
time_json = read_json('../data/time/time_day1.json')
time_json
```

```python
def to_csv(data):
    string = 'time,sid,change\n'
    for log in data:
        time = log['t']
        for sensor in log['s']:
            sid = sensor[0]
            change = sensor[1]
            string += '{},{},{}\n'.format(time,sid,change)
            
    with open('../data/logs/day1_all.csv', 'w') as f:
        f.write(string)
    
    return string

r = to_csv(time_json)
```
